#*Ejemplo redes neuronales*

```
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
import numpy as np
from matplotlib import pyplot as plt
```

```
data = load_iris()
X = data['data']
t = data['target']
#print(data)

plt.plot(X[:50,0],X[:50,1],'or')
plt.plot(X[50:100,0],X[50:100,1],'dg')
plt.plot(X[100:150,0],X[100:150,1],'sb')
```

```
model = MLPClassifier(solver='lbfgs',learning_rate_init=0.1,hidden_layer_sizes=(20,60,100,200))
```

```
Xtrain, Xtest, tTrain, tTest = train_test_split(X,t,test_size=0.2)
```

```
model.fit(Xtrain,tTrain)
Ws = model.coefs_
#print(Ws[0])
#print(Ws[1])

y_pred = model.predict(Xtest)
cMat = confusion_matrix(tTest,y_pred)
print('confusion\n',cMat,'\n','acc= ',accuracy_score(tTest,y_pred))
```

```
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt


db = loadmat('respiration.mat')
t = db['t']
t = t/np.max(t)
x = db['x']
x = x/np.max(x)

plt.plot(x,t, '-r')
```

```
from sklearn.neural_network import MLPRegressor

model = MLPRegressor(solver='adam', activation='relu',
                     hidden_layer_sizes=(1000,90))
model.fit(x,t)
y_est = model.predict(x)

plt.plot(x,t, '-r')
plt.plot(x, y_est, ':b')
```
